```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
from tensorflow.keras.activations import sigmoid
```

```python
X_train = np.array([[1.0], [2.0]], dtype=np.float32) 
Y_train = np.array([[300.0], [500.0]], dtype=np.float32) 
```

```python
fig, ax = plt.subplots(1, 1)
ax.scatter(X_train, Y_train, marker="x", c="b", label="Data points")
ax.legend(fontsize="medium")
ax.set_ylabel("Price in 1000$")
ax.set_xlabel("Area of house in 1000 sq.ft.")
plt.show()
```

```python
linear_layer = Dense(units=1, activation="linear")
```

```python
linear_layer.get_weights() # Empty as it's not instantiated yet
```

```python
a1 = linear_layer(X_train[0].reshape(1, 1))

w, b = linear_layer.get_weights()
print(w, b)
```

```python
# Settings weights and bias
set_w = np.array([[200]])
set_b = np.array([100])

linear_layer.set_weights([set_w, set_b])
print(linear_layer.get_weights())
```

```python
a1 = linear_layer(X_train[0].reshape(1, 1))
print(a1)
```

```python
# Predictions
pred = linear_layer(X_train)
plt.plot(X_train, pred)
plt.scatter(X_train, pred, marker="o", c="r")
plt.show()
```
