```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])   #featuresu
y_train = np.array([300.0, 500.0])   #target value
```

```python
def compute_cost(x_train, y_train, w, b) -> float:
    m: int = len(y_train)
    cost: float = 0

    for i in range(m):
        f_wb: float = x_train[i] * w + b
        cost += (f_wb - y_train[i]) ** 2

    return cost
```

```python
def dj_dw(x_train, w: float, b: float) -> float:
    m: int = len(x_train)
    res: float = 0

    for i in range(m):
        f_wb: float = w * x_train[i] + b
        res += (f_wb - y_train[i]) * x_train[i]

    return res / m

    
def dj_db(x_train, w: float, b: float) -> float:
    m: int = len(x_train)
    res: float = 0

    for i in range(m):
        f_wb: float = w * x_train[i] + b
        res += (f_wb - y_train[i])

    return res / m
```

```python
def gradient_descent(n: int, learning_rate: float) -> tuple[float, float, list[float], list[float]]:
    w: float = 0
    b: float = 0
    cost: list[float] = []
    w_values: list[float] = []

    for _ in range(n):
        temp_w: float = w - (learning_rate * dj_dw(x_train, w, b))
        temp_b: float = b - (learning_rate * dj_db(x_train, w, b))

        w = temp_w
        b = temp_b
        curr_cost = compute_cost(x_train, y_train, w, b)
        
        w_values.append(w)
        cost.append(curr_cost)

    return (w, b, w_values, cost)
```

```python
def plot_cost(x_train, cost) -> None:
    plt.plot(x_train, cost)
    plt.xlabel("W")
    plt.ylabel("J(w, b)")
    plt.legend("X vs Cost")
    plt.show()
```

```python
w, b, w_values, cost = gradient_descent(10000, 0.01)

print(w, b)
plot_cost(w_values, cost=cost)
```
