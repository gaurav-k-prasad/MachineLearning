```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300, 500])
```

```python
print("Size of data: ", len(x_train))
```

```python
plt.scatter(x_train, y_train, marker='x', c='g')

plt.title("Housing Prices")
plt.xlabel("size in 1000 sq feet")
plt.ylabel("Prices in 1000 dollars")

plt.show()
```

Let's say that the w = 100, b = 100 for f_wb(x(i)) = w*x(i) + b

```python
w = 100
b = 100

# calculate manually everything
f_wb_1 = w * x_train[0] + b
f_wb_2 = w * x_train[1] + b

print(f_wb_1, f_wb_2)
```

```python
def calculate_model_output(weight, bias, examples) -> None:
    m = examples.shape
    f_wb = np.zeros(m)

    for i in range(len(examples)):
        f_wb[i] = weight * examples[i] + bias;

    return f_wb
```

```python
temp_f_wb = calculate_model_output(weight=200, bias=100, examples=x_train)
print(temp_f_wb)

plt.plot(x_train, temp_f_wb, c='b', label="Our Prediction")
plt.scatter(x_train, y_train, marker='x',c='r', label="Actual Values")

plt.title("Housing Prices")
plt.ylabel("Prices in $1000")
plt.xlabel("Size in 1000 sq. ft.")
plt.legend()
plt.show()

# We can now use this model to calculate real entities
cost_1200sq = 200 * 1.2 + 100
print("Price: ", cost_1200sq)
```
