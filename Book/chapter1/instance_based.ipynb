```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = "https://github.com/ageron/data/raw/main/lifesat/lifesat.csv"
df = pd.read_csv(data)
df.head()
```

```python
df.plot(x="GDP per capita (USD)", y="Life satisfaction", kind="scatter")
plt.axis((23_000, 60_000, 5, 8))
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = df.iloc[:, 1].to_numpy().reshape(-1, 1)
y = df.iloc[:, -1].to_numpy().reshape(-1, 1)

model = LinearRegression()
model.fit(X, y)
```

```python
model.predict([[37_655.2]])
```

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=3)
model.fit(X, y)

model.predict([[37_655.2]])
```
